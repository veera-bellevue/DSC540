```python
#Calculation of Descriptive statistics from a Dataframe
people = ['Ann', 'Brandon', 'Chen', 'David', 'Emily', 'Farook', 'Gagan', 'Hamish', 'Imran', 'Joseph', 'Katherine', 'Lily']
age = [21,12,32,45,37,18,28,52,5,40,48,15]
weight = [55,35,77,68,70,60,72,69,18,65,82,48]
height = [160,135,170,165,173,168,175,159,105,171,155,158]
import pandas as pd

#Construct a dictionary with the available series data by using the following command
people_dict = {'People':people, 'Age':age, 'Weight':weight, 'Height':height}
people_df = pd.DataFrame(people_dict)
people_df

#Print no of rows and columns in the dataframe
print(people_df.shape)

#Print no of rows in a column
print(people_df['Age'].count())

#Print the total of a column
print(people_df['Age'].sum())

#Print the mean of a column
print(people_df['Age'].mean())

#print the median of a column
print(people_df['Height'].median())

#print the maximum weight 
print(people_df['Weight'].max())

#print standard deviation of the weights 
print(people_df['Weight'].std())


#To calculate the percentile, we can call a function from NumPy and pass on the column(a panda series). For example, to calculate 
#75th and 25th percentiles of age distribution and their difference(called the inter-quartile range)

pcnt_75 = np.percentile(people_df['Age'],75)
pcnt_25 = np.percentile(people_df['Age'],25)
print(pcnt_75)
print(pcnt_25)
print("Inter-quartile range :", pcnt_75-pcnt_25)

#Describe command to find detailed description of the dataframe (descriptive statistics)
print(people_df.describe())
```

```python
#User Defined funtions with the apply method
def categorize_sales(price):
    if price < 50:
        return "Low"
    elif price < 200:
        return "Medium"
    else:
        return "High"

df_sample=df[['Customer Name','State','Sales']].sample(n=100)
df_sample.head(10)

#Using the user defined functions
df_sample['Sales Price Category'] = df_sample['Sales'].apply(categorize_sales)
df_sample.head(10)

#Using the Python supplied function
df_sample['Customer Name Length'] = df_sample['Customer Name'].apply(len)
df_sample.head(10)

#Using lambda functions
#Display discounted sales price if the original sales price is greater than 200
df_sample['Discounted Price'] = df_sample['Sales'].apply(lambda x:0.85*x if x>200 else x)
df_sample.head(10)
```
