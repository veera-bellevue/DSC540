```python
#Loading the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


#Read the Adult income names text file
file_path = "C:\\Users\\raghu\\Downloads\\adult_income_names.txt"

#Create a list to hold the variable names in Adult income names text file
names=[]
# Read the text file line by line
with open(file_path, 'r') as file:
    for line in file:
        file.readline()
        var=line.split(":")[0]
        names.append(var)
names

#Add Income response variable to the names list
names.append('Income')
names

#Read the adult income dataset
file_path = "C:\\Users\\raghu\\Downloads\\adult_income_data.csv"

# Load the Excel file into a DataFrame
df = pd.read_csv(file_path)
print(df.head())


#Assign column names
df.columns=names
print(df.head())

#Find missing values 
print(df.isnull().sum())

df_subset = df[['age','education','occupation']]
df_subset.head()

```

```python

# Create a histogram for age with a bin size of 20
plt.hist(df_subset['age'],color='orange',edgecolor='k',bins=20)
plt.title(f'Histogram of age with Bin Size 20')
plt.xlabel('age')
plt.ylabel('Frequency')
plt.show()
```

```python
#function to remove whitespaces from the string variables
def strip_whitespace(s):
    return s.strip()
```

```python
#Using apply method strip whitespaces in the string based columns
#make a new column to hold the updated values, copy this new column to the old column and then drop the new columns

df_subset['education_stripped'] = df['education'].apply(strip_whitespace)
df_subset['education'] = df_subset['education_stripped']

df_subset['occupation_stripped'] = df['occupation'].apply(strip_whitespace)
df_subset['occupation'] = df_subset['occupation_stripped']

df_subset.drop(labels=['education_stripped','occupation_stripped'],axis=1,inplace=True)
df_subset.head()
```

```python
#find the number of people with age between 30 and 40 and find the number of people in that range
df_filter=(df_subset['age']>=30) & (df_subset['age']<=50) 
df_subset[df_filter].shape[0]
```

```python
#group by education to understand how the mean age is distrbuted 
byage = df_subset.groupby(['education'])
print("Mean age distribution based on education \n", byage.mean())
```

```python
#Summary statistics of age when the data is grouped by occupation
byoccupation_stats = df_subset.groupby('occupation').describe()['age']
print("Following profession has the oldest workers")
print(byoccupation_stats[byoccupation_stats['mean']==byoccupation_stats['mean'].max()]['mean'].index[0])
print("\n\n")

#75th percentile of the age
pcnt_75 = np.percentile(df_subset['age'],75)

print("75th percentile of age is ",pcnt_75)
# byoccupation_stats.index[0] = 
#dropping the rows where occupation column has unknown values
# df.drop(df[df['Name'] == 'Bob'].index, inplace=True)
# print(byoccupation_stats.index[0])
# byoccupation_stats.drop(byoccupation_stats[byoccupation_stats['occupation'] == '?'].index, inplace=True)
largest_workforce = byoccupation_stats[byoccupation_stats['75%'] > pcnt_75]
print("\nFollowing are the occupations having largest workforce more than the 75th percentile of age")
print(largest_workforce)
print("\nWe can see that the row at the top has no proper value for occupation, if we leave that and consider the next valid")
print("\nOccupation value, it is 'Exec-managerial'. This occupation has largest workforce more than the 75th percentile of age ")
#Plotting 
plt.figure(figsize=(15,8))
plt.barh(y=byoccupation_stats.index, width=byoccupation_stats['count'])
plt.yticks(fontsize=13)
plt.show()
```

From the above statistics we can see that oldest workers on average are in "Exec-managerial" occupation
